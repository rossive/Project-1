```python
LogReg.score(X_train, y_train)
```

```python
y_pred = LogReg.predict(X_test)
```

```python
cm = pd.DataFrame(
    confusion_matrix(y_test, y_pred),
    columns=['Predicted Not Ward', 'Predicted Ward'],
    index=['True Not Ward', 'True Ward']
)

cm
```
